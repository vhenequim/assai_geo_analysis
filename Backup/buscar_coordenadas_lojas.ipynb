```python
#endereços = df_lojas['Endereço']
```

```python
lista_lat = []
lista_lng = []
```

```python
for i in range(len(df_lojas)):
    pesquisa = df_lojas['Endereço'][i]
    pesquisa = pesquisa.replace(' ', '+')
    request = requests.get(f'https://maps.googleapis.com/maps/api/geocode/json?address={pesquisa}&key=AIzaSyB0ZvkUSKpIVHZN-RSEDt-13XXymaXTMoo')
    dados = request.json()
    coord = dados['results'][0]['geometry']['location']
    lista_lat.append(coord['lat'])
    lista_lng.append(coord['lng'])
```

```python
coords = {'LAT': lista_lat, 'LONG': lista_lng}
```

```python
df_lojas['LAT'] = lista_lat
df_lojas['LONG'] = lista_lng
```
